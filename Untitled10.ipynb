```python
import pandas as pd

df = pd.read_csv("/content/Student_Satisfaction_Survey.csv", encoding='latin1')
df.head()
```

```python
df.isnull().sum()
df.dropna(inplace=True)
# df['Rating'] = df['Rating'].astype(int) # This line caused a KeyError because 'Rating' column does not exist
print(df.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create 'Rating' column from 'Average/ Percentage'
df['Rating'] = df['Average/ Percentage'].apply(lambda x: float(x.split(' / ')[0])).astype(int)

sns.countplot(x='Rating', data=df)
plt.title("Student Satisfaction Rating Distribution")
plt.show()
```

```python
from textblob import TextBlob

df['Sentiment_Score'] = df['Questions'].apply(
    lambda x: TextBlob(str(x)).sentiment.polarity
)

df['Sentiment'] = df['Sentiment_Score'].apply(
    lambda x: 'Positive' if x > 0 else 'Negative' if x < 0 else 'Neutral'
)

df[['Questions', 'Sentiment']].head()
```

```python
df['Sentiment'].value_counts().plot(
    kind='pie', autopct='%1.1f%%', figsize=(5,5)
)
plt.title("Student Feedback Sentiment Analysis")
plt.ylabel("")
plt.show()
```

```python
df.groupby('Course Name ')['Rating'].mean().sort_values().plot(kind='barh')
plt.title("Average Satisfaction by Course Name")
plt.show()
```
